```python
import tensorflow as tf
import keras
from pickle import dump
```

```python
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications.vgg16 import preprocess_input
from keras.layers import Input,Flatten,Lambda,Dense
from keras.models import Model
from keras.models import Sequential
```

```python
in_layer = Input(shape=(224, 224, 3))
top_model=keras.applications.DenseNet121(include_top=False,weights="imagenet",input_tensor=in_layer)
x = Flatten()(top_model.output)
y =Dense(4096)(x)
model = Model(inputs=top_model.inputs, outputs=y)
model.summary()
```

```python
from os import listdir
def extract_features(directory):
    features=dict()
    i=0
    for name in listdir(directory):
        filename = directory + '/' + name
        image = load_img(filename, target_size=(224, 224))
        image = img_to_array(image)
        image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
        image = preprocess_input(image)
        feature = model.predict(image, verbose=0)
        image_id = name.split('.')[0]
        features[image_id] = feature
        print("Image "+str(i)+" features extracted")
        print(feature)
        print(feature.shape)
        i += 1
    return features
```

```python
directory = 'Flicker8k_Dataset'
```

```python
import time
start_time=time.time()
features = extract_features(directory)
end_time=time.time()
print("Extraction time: ")
print((end_time-start_time)/60)
print(len(features))
dump(features, open('densenetfeatures.pkl', 'wb'))
```
